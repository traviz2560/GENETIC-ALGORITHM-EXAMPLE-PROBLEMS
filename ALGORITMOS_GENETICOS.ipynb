#**Algoritmos Genéticos**

es una parte de la Computación Evolutiva (EC) que se inspira en el proceso de evolución y selección natural de los seres vivos.

Los algoritmos genéticos se utilizan generalmente para superar problemas de optimización y búsqueda. Este algoritmo es un algoritmo general, por lo que puede implementarse fácilmente en varios problemas y puede proporcionar mejores resultados para cada iteración de la solución de búsqueda. Los algoritmos genéticos pueden encontrar las mejores soluciones a partir de conjuntos de candidatos que son amplios y tienen muchos puntos óptimos y, además, los resultados tienden a ir al óptimo global en comparación con métodos similares, como la escalada , la búsqueda en profundidad , etc.

##**Términos en algoritmos genéticos**

* **Población**, recopilación de posibles soluciones.
* **Cromosoma**, una posible solución
* **Genotipo**, elementos contenidos en los cromosomas.
* **Fenotipo**, valor del genotipo

## **Crossover**

Crossover es un proceso para producir nuevos cromosomas. El cruce se realiza entre dos cromosomas, y el resultado del cruce son dos cromosomas.

Hay dos tipos de enfoques para determinar nuevos cromosomas, generacional y de estado estacionario.

* **Generacional**. Se crean cromosomas nuevos en la misma cantidad que los anteriores y, al final de la iteración, la población anterior se reemplaza por la nueva.
* **Estado estable**. A diferencia del método generacional la cantidad de cromosomas nuevos que se crean no es la misma que la de los anteriores, sino solo uno o dos. El nuevo cromosoma reemplazará al viejo cromosoma.

```python
import random

import numpy

import pandas

import math

from deap import algorithms
from deap import base
from deap import creator
from deap import tools
```

## **Problema 1**

Encontrar el valor de $x$, que maximiza la función:

$ f(x) = −x^2 + 2x + 8 $

$x$: es un número real en el intervalo $[0,10]$

```python
try:
  del creator.FitnessMax
  del creator.Individual
except:
  pass
  
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)
```

```python
toolbox = base.Toolbox()
toolbox.register("attr_real", random.uniform, 0, 10)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_real, 1)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
```

```python
def evalOneMax(individual):
  x = individual[0]
  return (-x**2+2*x+8,)
```

```python
toolbox.register("evaluate", evalOneMax)
toolbox.register("mate", tools.cxSimulatedBinaryBounded, eta=0.05, low=0, up=10)
toolbox.register("mutate", tools.mutGaussian, mu=0.05, sigma=0.005, indpb=0.05)
toolbox.register("select", tools.selTournament, tournsize=3)
```

```python
pop = toolbox.population(n=300)
hof = tools.HallOfFame(1)
stats = tools.Statistics(lambda ind: ind.fitness.values)
stats.register("avg", numpy.mean)
stats.register("std", numpy.std)
stats.register("min", numpy.min)
stats.register("max", numpy.max)

pop, log = algorithms.eaSimple(pop, toolbox, cxpb=0.5, mutpb=0.2, ngen=20, stats=stats, halloffame=hof, verbose=False)
```

```python
#@title

data = [[i for i in item.values()] for item in log]
df = pandas.DataFrame(data, columns=log.header)
df.style.hide_index()
```

```python
print("el valor que maximiza la función es: ",round(hof[0][0],5))
```

```python
try:
  del creator.FitnessMax
  del creator.Individual
except:
  pass
  
creator.create("FitnessMax", base.Fitness, weights=(-1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)
```

```python
toolbox = base.Toolbox()
toolbox.register("attr_real1", random.uniform, -10, 0)
toolbox.register("attr_real2", random.uniform, -6.5, 0)
toolbox.register("individual", tools.initCycle, creator.Individual, 
                 (toolbox.attr_real1, toolbox.attr_real2), 1)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
```

```python
def evalOneMax(individual):
  x = individual[0]
  y = individual[1]
  fxy = math.sin(y)*math.exp((1-math.cos(x))**2)+math.cos(x)*math.exp((1-math.sin(y))**2)+(x-y)**2
  return (fxy,)

def feasible(individual):
  x = individual[0]
  y = individual[1]
  if ((x+5)**2+(y+5)**2)<25:
    return True
  return False

def distance(individual):
  x = individual[0]
  y = individual[1]
  return (x+5)**3+(y+5)**3
```

```python
toolbox.register("evaluate", evalOneMax)
toolbox.decorate("evaluate", tools.DeltaPenalty(feasible, 10000, distance))
toolbox.register("mate", tools.cxSimulatedBinaryBounded, eta=0.1, low=[-10,-6.5], up=[0,0])
toolbox.register("mutate", tools.mutGaussian, mu=0.05, sigma=0.005, indpb=0.05)
toolbox.register("select", tools.selTournament, tournsize=3)
```

```python
pop = toolbox.population(n=400)
hof = tools.HallOfFame(1)
stats = tools.Statistics(lambda ind: ind.fitness.values)
stats.register("avg", numpy.mean)
stats.register("std", numpy.std)
stats.register("min", numpy.min)
stats.register("max", numpy.max)

pop, log = algorithms.eaSimple(pop, toolbox, cxpb=0.5, mutpb=0.5, ngen=20, stats=stats, halloffame=hof, verbose=False)
```

```python
toolbox = base.Toolbox()
toolbox.register("attr_real", random. uniform, 0, 0.25)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_real, 5)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
```

```python
def evalOneMax(individual):
  x11 = individual[0]
  x31 = individual[1]
  x12 = individual[2]
  x22 = individual[3]
  x32 = individual[4]
  U = (x11+x31)*5+(x12+x22+x32)*9
  return (U,)

def feasible(individual):
  x11 = individual[0]
  x31 = individual[1]
  x12 = individual[2]
  x22 = individual[3]
  x32 = individual[4]

  r1 = (x11*10+x12*15)<=(4.2*6)
  r2 = (x22*12)<=(0.5*6)
  r3 = (x31*4+x32*6)<=(1.8*6)

  if r1 and r2 and r3 and x11>0 and x31>0 and x12>0 and x22>0 and x32>0:
    return True
  return False
```

```python
toolbox = base.Toolbox()
toolbox.register("attr_int1", random.uniform, 0, 6000)
toolbox.register("attr_int2", random.uniform, 0, 9000)
toolbox.register("individual", tools.initCycle, creator.Individual, 
                 (toolbox.attr_int1, toolbox.attr_int2), 1)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
```

```python
def evalOneMax(individual):
  x1 = individual[0]
  x2 = individual[1]
  U = x1*3+x2*5
  return (U,)

def feasible(individual):
  x1 = individual[0]
  x2 = individual[1]

  r1 = (x1*1)<=(4000)
  r2 = (x2*2)<=(12000)
  r3 = (x1*3+x2*2)<=(18000)

  if r1 and r2 and r3 and x1>=0 and x2>=0:
    return True
  return False
```
